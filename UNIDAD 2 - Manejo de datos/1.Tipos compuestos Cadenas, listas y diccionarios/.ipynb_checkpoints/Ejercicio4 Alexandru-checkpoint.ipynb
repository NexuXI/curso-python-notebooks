## Ejercicio 4: Colecciones Colegio

#### Gestionar un colegio con 3 aulas.
   
#### Pintar menu:
    1. Insertar alumno. --> En que aula quieres meter al alumnos? (1,2,3) --> DNI, Nombre, Apellidos.
    2. Mostar alumnos colegio --> Mostar todos los alumnos del colegio por aula.
    3. Mostrar alumnos aula --> De que aula quieres mostrar los alumnos? (1,2,3).
    4. Buscar alumno --> Preguntar DNI --> En que aula se encuentra el alumno con ese DNI.
    5. Borrar alumno --> Preguntar DNI --> Borrar el alumno con el DNI.
    6. Salir

```python
 ## Crear  lista Colegio, con lista clases y estas con diccionarios alumnos
aula1 = []
aula2 = []
aula3 = []
colegio = [aula1, aula2, aula3]

## Crear funcion insertar alumnos
def InsertarAlumnos(colegi, aul, nif, nomb, apel):
    ## Restar 1 para tener el numero del aula dentro del colegio
    aul = aul-1
    ## Comprobar que el aula introducido no es mayor de 2 o menor de 0
    if not aul > 2 or aul < 0:
        ## Insertar alumno
        colegi[aul].append({"DNI": nif, "Nombre": nomb, "Apellidos": apel})
        ## Mensaje de confirmacion 
        print("El alumno: ", nomb, " " , apel , " con DNI: " , nif , " ha sido añadido correctamente al aula: ",(aul+1) ,"." )
    else:
        ## Mensaje de error
        print("El aula introducida no es correcta")
    
    ## Devuelve el colegio ahora con el alumno añadido
    return colegi


## Crear funcion ver todos los alumnos
def VerTodoAlumnos(colegi):
    ## Crear un contador para contar el numero del aula
    cont = 1
    ## Recorrer las aulas del colegio
    for aula in colegi:
        print("-------------------------")
        print("Alumnos del Aula " , cont , ".")
        print("-------------------------")
        ## Recorrer los alumnos de las aulas
        for alumno in aula:
            print(alumno["Nombre"], " " , alumno["Apellidos"], " con DNI: ", alumno["DNI"] )
        ## Sumar uno al contador
        cont+=1

        
## Crear funcion ver los alumnos de un aula
def VerAlumnosClase(colegi, aul):
    print("-------------------------")
    print("Alumnos del Aula " , aul, ".")
    print("-------------------------")
    ## Determinar cual es el aula especificando la posicion de esta en la lista del colegio
    aula = colegi[(aul-1)]
    ## Recorrer los alumnos de las aulas
    for alumno in aula:
        ## Mostrar los alumnos del aula
        print(alumno["Nombre"], " " , alumno["Apellidos"], " con DNI: ", alumno["DNI"])
    
    
## Crear funcion buscar alumno por DNI 
def BuscarAlumno(colegi, nif):
    ## Crear un contador para contar el numero del aula
    cont = 0
    ## Boleano para derterminar si ha encontrado al alumno
    encontrado = False
    ## Recorrer las aulas del colegio
    for aula in colegi:
        ## Recorrer los alumnos de las aulas
        for alumno in aula:
            ## Comprobar si el DNI coincide
            if alumno["DNI"] == nif:
                ## Cambia el booleano a True
                encontrado = True
                ## Guardar alumno y aula en una variable
                alumno_y_aula = [alumno, cont]
                ## Devuelve la variable con el alumno y el contador
                return alumno_y_aula
        ## Sumar uno al contador
        cont+=1
        
    if not encontrado:
        print("El alumno buscado no ha sido encontrado.")

    
## Crear funcion borrar alumno por DNI
def BorrarAlumno(colegi, nif):
    ## Crear un contador para contar el numero del aula
    cont = 0
    ## Crear un contador para contar el numero del alumno
    alumn = 0
    ## Boleano para derterminar si ha encontrado al alumno
    encontrado = False
    ## Recorrer las aulas del colegio
    for aula in colegi:
        ## Recorrer los alumnos de las aulas
        for alumno in aula:
            ## Comprobar si el DNI coincide
            if alumno["DNI"] == nif:
                ## Cambia el booleano a True
                encontrado = True
                ## Borra al alumno 
                colegi[cont].pop(alumn)
                ## Devuelve el colegio ahora con el alumno eliminado
                return colegi
            ## Sumar un al contador
            alumn+=1
        ## Sumar uno al contador
        cont+=1
        
    if not encontrado:
        ## Mensaje de error
        print("El alumno buscado no ha sido encontrado.")
    
    
## Declaramos opcion
opcion = 0
## Comienza el bucle, si la opcion es 5 sale
while (not opcion == 6):
    ## Imprime las opciones
    print("---------------------------------------")
    print("1. Insertar alumno.")
    print("2. Mostra alumnos del colegio.")
    print("3. Mostra alumnos del aula.")
    print("4. Buscar alumno por DNI.")
    print("5. Borrar alumno por DNI.")
    print("6. Salir.")
    print("---------------------------------------")
    
    ## Lee la opcion escrita
    opcion = int(input("Seleccione una opcion: "))
    
    ## Revisa si la opcion es 1
    if(opcion ==1):
        
        ## Pedir los datos para insertar
        aula = int(input("Escriba el numero del aula: 1, 2 o 3. "))
        dni = input("Escriba el DNI del Alumno. ")
        nombre = input("Escriba el Nombre del Alumno. ")
        apellidos = input("Escriba los Apellidos del Aulumno. ")
        ## Actualizar el colegio ahora con el alumno introducido
        colegio = InsertarAlumnos(colegio, aula, dni, nombre, apellidos)
        ## Actualizar las aulas
        aula1 = colegio[0]
        aula2 = colegio[1]
        aula3 = colegio[2]
        
        
    elif(opcion ==2):
        
        ## LLamar a la funcion que muestra todos los alumnos
        VerTodoAlumnos(colegio)
           
            
    elif(opcion ==3):
        
        ## Pedir los datos para insertar
        num_aula = int(input("Escriba el numero del aula: 1, 2 o 3. "))
        ## LLamar a la funcion que nos muestra los alumnos de un aula
        VerAlumnosClase(colegio, num_aula)
        
           
    elif(opcion ==4):
        
        ## Pedir los datos para insertar
        dni = input("Escriba el DNI del alumno a buscar. ")
        ## LLamar a la funcion que nos devuelve al alumno buscado
        alumno_y_aula = BuscarAlumno(colegio, dni)
        ## Combrueba que el resultado devuelto de la funcion  no es Null 
        if alumno_y_aula is None:
            print("Compruebe el DNI y vuelva a intentarlo. ")
        else:
            ## Guarda el alumno en una variable
            alumno = alumno_y_aula[0]
            ## Guarda el aula en una variable
            aula = alumno_y_aula[1]
            ## Mostrar el alumno buscado
            print(alumno["Nombre"], " " , alumno["Apellidos"], " con DNI: ", alumno["DNI"], " del aula: ", aula )

            
    elif(opcion ==5):
        
        ## Pedir los datos para insertar
        dni = input("Escriba el DNI del alumno a borrar. ")
        ## Actualiza el colegio con el devuelto por la funcion
        colegio = BorrarAlumno(colegio, dni)
        if not colegio is None:
            ## Mensaje de confirmacion
            print("El alumno ha sido borrado.")
        
        
    elif(opcion ==6):
        
        ## Mostrar mensaje
        print("Adios!")
    
    else:
        
        ## Mostrar mensaje de error
        print("Respuesta incorrecta.")
```
